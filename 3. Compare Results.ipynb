```python
import os
import pandas as pd
```

```python
folder_path = 'results'
classic_file_names = ["bio_classic.csv", "evolutionary_classic.csv", "human_classic.csv", "math_classic.csv", 
                      "physics_classic.csv", "swarm_classic.csv", "system_classic.csv"]
cec_file_names = ["bio_cec.csv", "evolutionary_cec.csv", "human_cec.csv", "math_cec.csv", 
                  "physics_cec.csv", "swarm_cec.csv", "system_cec.csv"]
```

```python
classic_df = []
cec_df = []
for file_name in classic_file_names:
    file_path = os.path.join(folder_path, file_name)
    df = pd.read_csv(file_path)
    classic_df.append(df)

for file_name in cec_file_names:
    file_path = os.path.join(folder_path, file_name)
    df = pd.read_csv(file_path)
    cec_df.append(df)
```

```python
combined_classic_df = pd.concat(classic_df, ignore_index=True)
combined_cec_df = pd.concat(cec_df, ignore_index=True)
```
